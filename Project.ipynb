# 1. Paper Summary


In “Market Timing & Trading Strategies Using Asset Rotation” by Panagiotis Schizas and Dimitrios Thomakos (2010), the authors demonstrate that a market timing strategy based on rotation between two risky assets can generate statistically and economically significant performance gains over standard benchmarks. The paper argues that **relative returns** and **relative volatilities** are predictable and that such predictability can be systematically exploited through trading rules. Using Exchange Traded Fund (ETF) data, the authors compare rotation strategies with buy-and-hold and equally weighted portfolios. They show that even simple rotation models can outperform passive benchmarks, while more sophisticated specifications further enhance performance. Overall, the paper claims that asset rotation provides a practical and profitable framework for market timing strategies.
<br><br>

The paper’s **first main claim** is that **asset rotation between two risky assets is a viable market timing strategy**. Instead of switching between a risky and a risk-free asset, the authors rotate fully between two risky ETFs based on forecasts of relative performance. The strategy always keeps capital invested in the market but shifts exposure toward the asset expected to outperform. Empirical results across three ETF pairs show that this approach frequently outperforms both buy-and-hold and equally weighted portfolios. This demonstrates that relative asset selection can be more effective than static allocation. Overall, this point establishes asset rotation as a meaningful alternative to traditional market timing.
<br><br>

**The second main contribution** is that **relative returns and relative volatilities are predictable using simple econometric models**. The authors apply several regression-based specifications, including naïve mean models, moving average error structures, piecewise linear models with asymmetric effects, and autoregressive volatility models. Volatility is measured using a weekly realized range-based estimator constructed from daily high–low prices. Using rolling estimation windows, the models generate one-week-ahead forecasts for trading decisions. The results show that sign prediction rates often exceed 50% and are statistically significant in several cases. This confirms that relative performance between assets contains exploitable predictive information.
<br><br>

**The third main point** is that **forecast-based trading rules can convert statistical predictability into profitable trading outcomes**. The strategy follows a simple binary rule: at each period, all capital is allocated to the asset with the higher predicted relative return or lower predicted relative volatility. Performance is evaluated using cumulative wealth, average return, volatility, Sharpe ratio, and maximum loss. Across three trading days and three asset pairs, rotation strategies outperform benchmarks in most cases. Even the naïve moving-average model performs competitively, showing that complex modeling is not required for profitability. This point demonstrates the operational simplicity and effectiveness of the rotation methodology.
<br><br>

The paper further argues that **the success of rotation strategies depends heavily on sign prediction accuracy and volatility behavior**. Strategies with higher sign success ratios consistently achieve better trading performance. Bootstrap tests confirm that sign prediction rates above 50% are statistically significant in the most successful cases. The authors also examine whether volatility is related to correct sign predictions using chi-square independence tests. Although results vary across asset pairs, volatility dynamics appear to influence forecasting success in several settings. This section highlights that directional accuracy, rather than return magnitude alone, is central to profitable trading.
<br><br>

Taken together, the paper shows that asset rotation strategies based on forecasting relative returns and volatilities can outperform traditional passive investment approaches. The findings support the market timing literature and challenge the strict form of market efficiency. These results are important because they provide a replicable methodology for active trading based on statistical predictability. For replication and further research, the paper offers a well-structured foundation for extending asset rotation strategies to other assets, time periods, and model specifications.

# 2. Summary of Hypotheses and Tests

The original paper does not explicitly enumerate its hypotheses in a formal list; instead, the research hypotheses are embedded in the empirical design and testing framework. These hypotheses can be summarized into three closely related research propositions, together with their corresponding empirical tests.

### Hypothesis 1: Economic Profitability of Asset Rotation Strategies  

**Subject:** The economic performance of forecast-based asset rotation strategies.

**Dependent variables:** Cumulative portfolio wealth, average return, volatility, Sharpe ratio, and maximum drawdown.

**Independent variables:** Trading signals derived from relative return and relative volatility forecasts.

**Anticipated outcome:** Asset rotation strategies outperform passive benchmark portfolios in both absolute and risk-adjusted performance after transaction costs.

**Validation / Refutation:** This hypothesis is validated if rotation strategies achieve higher cumulative wealth and Sharpe ratios than buy-and-hold and equally weighted benchmarks. It is refuted if such outperformance disappears.
<br><br>

### Hypothesis 2: Predictability of Relative Returns and Volatility  

**Subject:** The predictability of relative returns and relative volatilities between two risky assets.

**Dependent variables:** Realized relative return and realized relative volatility.

**Independent variables:** Forecasts generated from regression-based relative return models and autoregressive relative volatility models estimated using rolling windows of historical data.

**Anticipated outcome:** Directional prediction accuracy exceeds random chance (greater than 50%), and volatility forecasts are statistically associated with realized volatility.

**Validation / Refutation:** This hypothesis is validated if the sign success ratio $SR = \frac{\text{Number of correct sign forecasts}}{\text{Total number of forecasts}}\quad $
is statistically greater than 0.5 and if volatility forecast regressions produce statistically significant slope coefficients. It is refuted if forecasting accuracy is statistically indistinguishable from random behavior.
<br><br>

### Hypothesis 3: Volatility Dependence of Directional Forecasting Success  

**Subject:** The relationship between volatility regimes and directional forecasting accuracy.

**Dependent variables:** Binary indicators of predicted positive relative returns and correctly predicted positive relative returns.

**Independent variable:** Volatility regime, defined by relative volatility dominance between the two assets.

**Anticipated outcome:** Directional forecasting success is systematically related to volatility conditions.

**Validation / Refutation:** This hypothesis is validated if chi-square independence tests following Kavajecz and Odders-White (2001) reject the null hypothesis of independence between directional prediction success and volatility regimes at conventional significance levels. It is refuted if directional forecasting accuracy is statistically independent of volatility conditions.

### Extension: Temporal Robustness  
This extension examines whether the economic profitability and statistical properties of asset rotation strategies remain stable when the sample period is extended to 2025. While the original study ends in 2008, the extended sample spans multiple distinct market regimes, including post-crisis recovery, prolonged quantitative easing, the COVID-19 shock, and subsequent normalization phases. The analysis focuses on the time-stability of forecasting performance and trading profitability, with portfolio returns, Sharpe ratios, drawdowns, and directional accuracy as dependent variables, and forecast-based allocation rules as independent variables.

This extension evaluates whether the documented profitability reflects a persistent structural characteristic of asset rotation strategies or is primarily driven by specific historical conditions.

**Validation / Refutation:** This extension is validated if performance remains statistically and economically significant in the extended sample period and is refuted if results deteriorate materially over time.

## 4.2 Data Preprocessing

After cleaning, each dataset is transformed into a standardized daily time series indexed by trading date. Using the cleaned daily data, I compute the Parkinson range-based variance estimator

$$
\sigma^2_{d,t} = \frac{(\ln(H_t) - \ln(L_t))^2}{4\ln(2)},
$$
where $H_{s}$ and $L_{s}$ denote the sth day high and low prices,
and aggregate daily variances into weekly realized volatility $
V_{t,i} = \sum_{s=1}^{5} \sigma_{st,i}.
$
<br><br>
Weekly prices are constructed by selecting closing prices on Monday, Wednesday, or Friday and aligning them with the corresponding volatility measures.
<br><br>
This data processing procedure produces consistent daily and weekly datasets for all ETFs and supports robustness analysis under alternative week-ending conventions.

```python
import numpy as np
import pandas as pd

TICKERS = ["SPY", "XLF", "QQQ", "OIH"]

def add_week_end_label(df_daily, week_end_day):
    target = {"MON": 0, "WED": 2, "FRI": 4}[week_end_day]
    idx = df_daily.index
    days_ahead = (target - idx.weekday) % 7
    week_end = (idx + pd.to_timedelta(days_ahead, unit="D")).normalize()
    out = df_daily.copy()
    out["week_end"] = week_end
    return out

def parkinson_sigma_daily(df_daily):
    hl = np.log(df_daily["High"]) - np.log(df_daily["Low"])
    return (hl ** 2) / (4.0 * np.log(2.0))

def make_weekly_from_daily(df_daily, week_end_day):
    d = df_daily[["Open","High","Low","Close"]].copy().sort_index()
    d["sigma_d"] = parkinson_sigma_daily(d) /2
    d["V_week"] = d["sigma_d"].rolling(5, min_periods=5).sum()

    target = {"MON": 0, "WED": 2, "FRI": 4}[week_end_day]
    weekly = d[d.index.weekday == target][["Close","V_week"]].copy()
    weekly.index = weekly.index.normalize()
    return weekly.dropna().sort_index()

def load_weekly_data_from_csv_dict(data_dict, tickers=TICKERS):
    daily = {t: data_dict[t].copy() for t in tickers}
    weekly = {"MON": {}, "WED": {}, "FRI": {}}
    for day in weekly:
        for t in tickers:
            weekly[day][t] = make_weekly_from_daily(daily[t], day)
    return daily, weekly

daily, weekly = load_weekly_data_from_csv_dict(data, TICKERS)
print(weekly["WED"]["XLF"].head())
```

## 4.3 Key Variable Definition and Calculation

Variables used in the empirical analysis are summarized following Schizas and Thomakos (2010).

*   **Weekly return**
$$
R_{t,i} = \log(C_{t,i}) - \log(C_{t-1,i})
$$
where $C_{t,i}$ is the weekly closing price of ETF i at week t.

*  **Relative return**
$$
y_t = R_{t,i} - R_{t,j}
$$

* **Relative volatility in level and logarithmic**
$$
V_t = V_{t,i} - V_{t,j}
$$

$$
v_t = \log(V_{t,i}) - \log(V_{t,j})
$$

```python
def build_pair_variables(weekly_dict, i, j):
    I = weekly_dict[i].copy()
    J = weekly_dict[j].copy()

    idx = I.index.intersection(J.index)
    df = pd.DataFrame(index=idx)

    df["C_i"] = I.loc[idx, "Close"]
    df["C_j"] = J.loc[idx, "Close"]
    df["R_i"] = np.log(df["C_i"]).diff()
    df["R_j"] = np.log(df["C_j"]).diff()

    df["V_i"] = I.loc[idx, "V_week"]
    df["V_j"] = J.loc[idx, "V_week"]

    df["y"] = df["R_i"] - df["R_j"]
    df["V"] = df["V_i"] - df["V_j"]
    df["v"] = np.log(df["V_i"]) - np.log(df["V_j"])

    return df.dropna()

print(build_pair_variables(weekly["WED"], "SPY", "XLF").head())
```

# 5. Model Construction

Following Schizas and Thomakos (2010), I implement five alternative forecasting models to generate relative return and volatility signals for asset rotation: (i) Naïve Mean, (ii) ARIMA, (iii) Difference of Log-Volatility, (iv) Log-Volatility Ratio, and (v) Piecewise Linear Regression. All models are estimated using rolling windows and evaluated out-of-sample.

## 5.1 Naïve Mean Model

The Naïve model serves as a benchmark and assumes that future relative returns equal their historical mean:

$$
y_t = \beta_0 + u_t.
$$
Forecasts are constructed as:

$$
\hat y_t = \frac{1}{n_0}\sum_{k=1}^{n_0} y_{t-k}.
$$

This model ignores dynamics and explanatory variables and corresponds to Equation (7) in the original study.

---

## 5.2 ARIMA (Moving Average) Model

To capture serial dependence in forecast errors, I implement a moving average model:

$$
y_t = \beta_0 + u_t, \quad
u_t = \varepsilon_t + \sum_{k=1}^{q}\theta_k \varepsilon_{t-k}.
$$

This specification is equivalent to an ARIMA(0,0,q) model and corresponds to Equation (8). The lag order $q$ is fixed across rolling windows.

---

## 5.3 Difference of Log-Volatility Model

Relative volatility is defined as the difference between log-volatilities:

$$
v_t = \log(V_{i,t}) - \log(V_{j,t}).
$$

I estimate an autoregressive model:

$$
v_t = \alpha_0 + \sum_{k=1}^{p} \alpha_k v_{t-k} + w_t.
$$

The lag length $p$ is selected using the Akaike Information Criterion (AIC).

---

## 5.4 Log-Volatility Ratio Model

In this specification, I estimate separate autoregressive models for individual asset volatilities:
$$
\log(V_{i,t}) = \phi_0 + \sum_{k=1}^{p_i} \phi_k \log(V_{i,t-k}) + \eta_t,
$$

$$
\log(V_{j,t}) = \psi_0 + \sum_{k=1}^{p_j} \psi_k \log(V_{j,t-k}) + \xi_t.
$$

Relative volatility forecasts are obtained as:
$$
\hat v_t = {\log(\hat V_{i,t})} - {\log(\hat V_{j,t})}.
$$

Lag orders are selected using AIC. This approach corresponds to Equation (10) in the original study.

---

## 5.5 Piecewise Linear Regression Model

The main forecasting model extends the baseline by incorporating asymmetric and regime-dependent effects:

$$
\begin{aligned}
y_t = &\ \beta_0 + \beta_1 y_{t-1} + \beta_2 V_{t-1}
+ \beta_3 I_t^y + \beta_4 I_t^V \\
&+ \beta_5 y_{t-1} I_t^y + \beta_6 V_{t-1} I_t^V
+ \beta_7 y_{t-1} I_t^V + \beta_8 V_{t-1} I_t^y + u_t,
\end{aligned}
$$

where:

$$
I_t^y = \mathbf{1}(y_{t-1}<0), \quad
I_t^V = \mathbf{1}(V_{t-1}<c).
$$

## 5.6 Implementation

At each time t, model parameters are estimated using a rolling window of $n_0 = 104$ weekly observations. For the pairs SPY–OIH, SPY–XLF, and SPY–QQQ, the out-of-sample evaluation windows differ for Mondays, Wednesdays, and Fridays, reflecting differences in data availability.

The volatility threshold c is fixed, selecting from in-sample quantiles. This model captures asymmetric responses under different return-volatility regimes.

Since the forecasting models produce $\hat y_{t|t-1}$, I apply a one-period forward shift to obtain $\hat y_{t+1|t}$, which is then used to determine portfolio positions and realized returns.

```python
# Naive
def forecast_naive_mean(df, n0=104, n1=200):
    y = df["y"].astype(float)
    yhat = pd.Series(index=y.index, dtype=float)
    start = n0
    end = start + n1
    for t in range(start, min(end, len(y))):
        yhat.iloc[t] = y.iloc[t-n0:t].mean()
    return yhat.shift(-1)

# ARIMA
def forecast_arima_ma(df, q=1, n0=104, n1=200):
    y = df["y"].astype(float).to_numpy()
    idx = df.index
    yhat = pd.Series(index=idx, dtype=float)
    start = n0
    end = min(start + n1, len(y))
    for t in range(start, end):
        train = y[t-n0:t]
        train = train[~np.isnan(train)]
        if len(train) < max(10, q+2):
            yhat.iloc[t] = np.nan
            continue
        try:
            res = ARIMA(train, order=(0,0,q), trend="c").fit()
            yhat.iloc[t] = float(res.forecast(1)[0])
        except Exception:
            yhat.iloc[t] = float(np.mean(train))
    return yhat.shift(-1)

# Diff Log-Vol
def forecast_diff_logvol_ar(df, n0=104, n1=200, pmax=20):
    v = df["v"].replace([np.inf, -np.inf], np.nan).astype(float)
    yhat = pd.Series(index=v.index, dtype=float)
    start = n0
    end = start + n1
    for t in range(start, min(end, len(v))):
        train = v.iloc[t-n0:t].dropna()
        if len(train) < max(30, n0//2):
            yhat.iloc[t] = train.mean() if len(train) else np.nan
            continue
        p = select_ar_order_aic(train, pmax=pmax)
        try:
            res = AutoReg(train, lags=p, trend="c", old_names=False).fit()
            yhat.iloc[t] = float(res.predict(start=len(train), end=len(train)).iloc[0])
        except Exception:
            yhat.iloc[t] = train.mean()
    return yhat.shift(-1)

# Log-Vol Ratio
def forecast_logvol_ratio_ar(df, n0=104, n1=200, pmax=20):
    """
    Return yhat := forecasted (logVi - logVj). trading rule uses yhat > 0.
    """
    logVi = np.log(df["V_i"]).replace([np.inf, -np.inf], np.nan).astype(float)
    logVj = np.log(df["V_j"]).replace([np.inf, -np.inf], np.nan).astype(float)
    yhat = pd.Series(index=df.index, dtype=float)
    start = n0
    end = start + n1
    for t in range(start, min(end, len(df))):
        tr_i = logVi.iloc[t-n0:t].dropna()
        tr_j = logVj.iloc[t-n0:t].dropna()
        if len(tr_i) < max(30, n0//2) or len(tr_j) < max(30, n0//2):
            yhat.iloc[t] = (tr_i.mean() - tr_j.mean())
            continue
        try:
            p_i = select_ar_order_aic(tr_i, pmax=pmax)
            p_j = select_ar_order_aic(tr_j, pmax=pmax)

            res_i = AutoReg(tr_i, lags=p_i, trend="c", old_names=False).fit()
            res_j = AutoReg(tr_j, lags=p_j, trend="c", old_names=False).fit()

            f_i = float(res_i.predict(start=len(tr_i), end=len(tr_i)).iloc[0])
            f_j = float(res_j.predict(start=len(tr_j), end=len(tr_j)).iloc[0])

            yhat.iloc[t] = f_i - f_j
        except Exception:
            yhat.iloc[t] = (tr_i.mean() - tr_j.mean())
    return yhat.shift(-1)

def select_ar_order_aic(series: pd.Series, pmax: int = 20) -> int:
    s = series.dropna().astype(float)
    n = len(s)
    best_p = 1
    best_aic = np.inf
    for p in range(1, pmax + 1):
        try:
            res = AutoReg(s, lags=p, trend="c", old_names=False).fit()
            if res.aic < best_aic:
                best_aic, best_p = res.aic, p
        except Exception:
            continue
    return best_p

# Piecewise linear
def build_piecewise_features(df, c):
    d = df.copy()
    d["y_l1"] = d["y"].shift(1)
    d["V_l1"] = d["V"].shift(1)

    d["Iy"] = (d["y_l1"] < 0).astype(int)     # I(y_{t-1} < 0)
    d["IV"] = (d["V_l1"] < c).astype(int)     # I(V_{t-1} < c)

    d["yIy"] = d["y_l1"] * d["Iy"]            # y_{t-1} * I^y
    d["VIV"] = d["V_l1"] * d["IV"]            # V_{t-1} * I^V
    d["yIV"] = d["y_l1"] * d["IV"]            # y_{t-1} * I^V
    d["VIy"] = d["V_l1"] * d["Iy"]            # V_{t-1} * I^y

    cols = ["y_l1", "V_l1", "Iy", "IV", "yIy", "VIV", "yIV", "VIy"]

    d = d.dropna(subset=cols)
    X = sm.add_constant(d[cols], has_constant="add")
    y = d["y"]
    return d, X, y

def choose_c_fixed(df, train_end=104, q=0.5):
    v_train = df["V"].iloc[:train_end].dropna()
    return float(v_train.quantile(q))

def forecast_piecewise_linear(df, c, n0=104, n1=200, min_train=30):
    yhat = pd.Series(index=df.index, dtype=float)
    start = n0
    end = min(n0 + n1, len(df))
    for t in range(start, end):
        win_plus = df.iloc[t - n0 : t + 1].copy()
        d2, X2, y2 = build_piecewise_features(win_plus, c)
        if len(X2) < 2:
            yhat.iloc[t] = np.nan
            continue
        X_tr, y_tr = X2.iloc[:-1], y2.iloc[:-1]
        X_te = X2.iloc[-1:]
        mask = ~y_tr.isna()
        if mask.sum() < min_train:
            yhat.iloc[t] = float(y_tr[mask].mean()) if mask.sum() else np.nan
            continue
        try:
            res = sm.OLS(y_tr[mask], X_tr.loc[mask]).fit()
            yhat.iloc[t] = float(res.predict(X_te).iloc[0])
        except Exception:
            yhat.iloc[t] = float(y_tr[mask].mean())
    return yhat.shift(-1)
```

# 6. Trading Rules and Behavior

## 6.1 Trading Rules

The rotation trading strategy is based on one-period-ahead forecasts generated by the five forecasting models described in Section 5. For each model, I implement a fully-invested **binary asset rotation strategy**.

For **return-based models**, the strategy holds ETF i if the predicted relative return is positive and ETF j otherwise:
$$
\text{Position   }_{\  i,t}^{(m)} =
\begin{cases}
1, & \hat{y}^{(m)}_{t|t-1} > 0, \\
0, & \hat{y}^{(m)}_{t|t-1} \le 0.
\end{cases}
$$

For **volatility-based models**, the same rule is applied using forecasts of relative log-volatility.

All available capital is fully invested in a single asset at each period. A portfolio switch occurs at time t only if the current signal differs from the signal at time t-1, which limits unnecessary turnover.

To avoid look-ahead bias, portfolio positions at time t are determined using forecasts generated at time t-1. The **realized strategy return** at time t is therefore computed as:

$$
R^{(m)}_t =
\mathbf{1}(\hat{y}^{(m)}_{t|t-1} > 0) R_{t,i}
+ \mathbf{1}(\hat{y}^{(m)}_{t|t-1} < 0) R_{t,j}.
$$

In addition, **directional forecasting accuracy** is measured by comparing the sign of the forecasted relative return with the realized relative return. Specifically, a correct sign prediction at time $t+1$ is recorded as
$$
S^{(m)}_{t+1}
=
\mathbf{1}(\hat y^{(m)}_{t+1|t} > 0)\mathbf{1}(y_{t+1} > 0)
+
\mathbf{1}(\hat y^{(m)}_{t+1|t} < 0)\mathbf{1}(y_{t+1} < 0).
$$

## 6.2 Implementation

This implementation closely follows the original study and replicates its key analytical techniques, including rolling-window estimation, one-step-ahead forecasting, binary rotation rules, and out-of-sample performance evaluation.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.ar_model import AutoReg

N1_TABLE = {
    "MON": {
        ("SPY","OIH"): 232,
        ("SPY","XLF"): 331,
        ("SPY","QQQ"): 322,
    },
    "WED": {
        ("SPY","OIH"): 264,
        ("SPY","XLF"): 374,
        ("SPY","QQQ"): 363,
    },
    "FRI": {
        ("SPY","OIH"): 256,
        ("SPY","XLF"): 331,
        ("SPY","QQQ"): 353,
    }
}

def perf_stats(simple_r: pd.Series, scale_sharpe=False) -> dict:
    r = simple_r.dropna()
    end_wealth = float((1 + r).prod())
    mean = float(r.mean())
    std = float(r.std(ddof=1))
    sharpe = mean / std if std > 0 else np.nan
    if scale_sharpe:
        sharpe = sharpe * 100
    min_r = float(r.min())
    return {
        "End wealth": end_wealth,
        "Average Return": mean * 100,       # weekly %
        "Standard Deviation": std * 100,    # weekly %
        "Sharpe Ratio": sharpe,             # weekly Sharpe
        "Minimum Realized Return": min_r * 100
    }

def positions_from_yhat(df: pd.DataFrame, yhat: pd.Series) -> pd.Series:
    s = yhat.reindex(df.index).astype(float)
    return (s > 0).astype(int).rename("pos_i_t")

def strategy_returns_from_yhat(df: pd.DataFrame, yhat: pd.Series, realized_col: str = "y") -> pd.DataFrame:
    d = df.copy()
    d["pos_i_t"] = positions_from_yhat(df, yhat)
    d["R_strat"] = d["pos_i_t"] * d["R_i"].shift(-1) + (1 - d["pos_i_t"]) * d["R_j"].shift(-1)
    y_next = d["y"].shift(-1)
    d["S_correct"] = (
        ((d["pos_i_t"] == 1) & (y_next > 0)) |
        ((d["pos_i_t"] == 0) & (y_next < 0))
    ).astype(int)
    return d[["R_strat", "S_correct", "pos_i_t"]]

PAIR_SAMPLE = {
    ("SPY","OIH"): ("2001-02-07", "2008-04-04"),
    ("SPY","XLF"): ("1998-12-22", "2008-04-04"),
    ("SPY","QQQ"): ("1999-03-10", "2008-04-04"),
}

def slice_pair_period(df, pair):
    start, end = PAIR_SAMPLE[pair]
    d = df.loc[start:end].copy()
    return d

def run_backtest_model(df_pair, model_func, n0, n1, scale_sharpe=False, **kwargs):
    yhat = model_func(df_pair, n0=n0, n1=n1, **kwargs)
    strat = strategy_returns_from_yhat(df_pair, yhat)

    strat_eval = strat.iloc[n0+1 : n0+n1+1].dropna(subset=["R_strat"])
    return perf_stats(strat_eval["R_strat"], scale_sharpe=scale_sharpe)

def bench_stats(df_pair, n0, n1):
    sl = df_pair.iloc[n0+1 : n0+n1+1]
    out_i  = perf_stats(sl["R_i"])
    out_j  = perf_stats(sl["R_j"])
    out_ew = perf_stats(0.5 * sl["R_i"] + 0.5 * sl["R_j"])
    return out_i, out_j, out_ew

def replicate_table_for_day(weekly_by_day, day):
    rows = []
    for pair in [("SPY","OIH"), ("SPY","XLF"), ("SPY","QQQ")]:
        i, j = pair
        scale = (j in ["XLF", "QQQ"])
        df = build_pair_variables(weekly_by_day, i, j)
        df = slice_pair_period(df, pair)
        n0 = 104
        n1 = N1_TABLE[day][pair]

        # (7) naive
        res7 = run_backtest_model(df, forecast_naive_mean,n0=n0, n1=n1,scale_sharpe=scale)

        # (8) ARIMA(0,0,q)
        res8 = run_backtest_model(df, forecast_arima_ma, n0=n0, n1=n1, q=1, scale_sharpe=scale)

        # (9) piecewise linear
        c = choose_c_fixed(df, train_end=104, q=0.3)
        res9 = run_backtest_model(df, forecast_piecewise_linear, n0=n0, n1=n1, c=c, scale_sharpe=scale)

        # (10) volatility models
        res10_diff  = run_backtest_model(df, forecast_diff_logvol_ar, n0=n0, n1=n1, pmax=18, scale_sharpe=scale)
        res10_ratio = run_backtest_model(df, forecast_logvol_ratio_ar, n0=n0, n1=n1, pmax=18, scale_sharpe=scale)

        # Benchmarks
        bh_i, bh_j, ew = bench_stats(df, n0=n0, n1=n1)

        rows.append({
            "Pair": f"{i}-{j}",
            "Naïve": res7,
            "ARIMA": res8,
            "Differ. of Volatility": res10_diff,
            "Volatility Ratio": res10_ratio,
            "Piecewise Linear": res9,
            "S&P500" if i=="SPY" else i: bh_i,
            j: bh_j,
            "Equally Weighted": ew,
        })
    return rows
```

## 6.3.3 Trading Behavior and Prediction Quality Analysis

To examine the trading behavior of the rotation strategies, I compute statistics on the number of trades, trading frequency, holding durations, and transition probabilities. These measures are calculated for all asset pairs and forecasting models, and representative results for Wednesdays are reported in Tables 4 to 6.

This analysis provides insights into the stability of trading signals and the degree of portfolio turnover across different forecasting models, which helps explain performance differences and transaction cost implications.

```python
import numpy as np
import pandas as pd

def eval_index(df_pair, n0, n1):
    return df_pair.index[n0+1 : n0+n1+1]

def trade_indicator_from_pos(pos: pd.Series) -> pd.Series:
    trade = (pos != pos.shift(1)).astype(int)
    if len(trade) > 0:
        trade.iloc[0] = 0
    return trade.rename("trade")

def transition_probs(trade: pd.Series) -> dict:
    x = trade.dropna().astype(int).values
    if len(x) < 2:
        return {
            "No Trade To No Trade": np.nan,
            "No Trade To Trade": np.nan,
            "Trade To No Trade": np.nan,
            "Trade To Trade": np.nan,
        }
    prev, nxt = x[:-1], x[1:]

    def p(a, b):
        mask = (prev == a)
        denom = mask.sum()
        return float((nxt[mask] == b).mean()) if denom > 0 else np.nan

    return {
        "No Trade To No Trade": p(0, 0),
        "No Trade To Trade":    p(0, 1),
        "Trade To No Trade":    p(1, 0),
        "Trade To Trade":       p(1, 1),
    }

def run_lengths(binary: np.ndarray, val: int) -> list[int]:
    lens, cur = [], 0
    for v in binary:
        if v == val:
            cur += 1
        else:
            if cur > 0:
                lens.append(cur)
                cur = 0
    if cur > 0:
        lens.append(cur)
    return lens

def duration_stats(trade: pd.Series) -> dict:
    x = trade.dropna().astype(int).values
    nt = run_lengths(x, 0)
    tr = run_lengths(x, 1)
    return {
        "Mean Duration No Trade": float(np.mean(nt)) if len(nt) else np.nan,
        "Max Duration No Trade":  float(np.max(nt))  if len(nt) else np.nan,
        "Mean Duration Trade":    float(np.mean(tr)) if len(tr) else np.nan,
        "Max Duration Trade":     float(np.max(tr))  if len(tr) else np.nan,
    }

def behavior_table_one_pair(
    weekly_by_day, pair, day="WED", n0=104, n1=None, q_ma=1, c_q=0.5,
    slice_func=slice_pair_period
):
    i, j = pair
    df = build_pair_variables(weekly_by_day, i, j)
    df = slice_func(df, pair)

    n1 = N1_TABLE[day][pair] if n1 is None else n1
    ev = eval_index(df, n0, n1)

    yhat_naive = forecast_naive_mean(df, n0=n0, n1=n1).reindex(df.index)
    yhat_ma    = forecast_arima_ma(df, q=q_ma, n0=n0, n1=n1).reindex(df.index)

    c = choose_c_fixed(df, train_end=n0, q=c_q)
    yhat_pw = forecast_piecewise_linear(df, c=c, n0=n0, n1=n1).reindex(df.index)

    yhat_map = {
        "Naïve": yhat_naive,
        "Mov. Avg.": yhat_ma,
        "PieceWise Linear": yhat_pw,
    }

    rows = []
    for name, yhat in yhat_map.items():
        pos_all = positions_from_yhat(df, yhat)
        pos = pos_all.reindex(ev).dropna()
        trade = trade_indicator_from_pos(pos).dropna()
        stats = {
            "Number of trades": int(trade.sum()),
            "Mean trading time": float(trade.mean()),
        }
        stats.update(transition_probs(trade))
        stats.update(duration_stats(trade))
        rows.append((name, stats))

    res = pd.DataFrame({k: v for k, v in rows}).T[
        ["Number of trades",
         "Mean trading time",
         "No Trade To No Trade",
         "No Trade To Trade",
         "Trade To No Trade",
         "Trade To Trade",
         "Mean Duration No Trade",
         "Max Duration No Trade",
         "Mean Duration Trade",
         "Max Duration Trade"]
    ]
    return res

def pretty(df: pd.DataFrame) -> pd.DataFrame:
    d = df.copy()
    d["Number of trades"] = d["Number of trades"].astype(int)
    for c in d.columns:
        if c != "Number of trades":
            d[c] = d[c].astype(float).round(3)
    return d

table4 = pretty(behavior_table_one_pair(weekly["WED"], ("SPY","OIH"), day="WED"))
table5 = pretty(behavior_table_one_pair(weekly["WED"], ("SPY","XLF"), day="WED"))
table6 = pretty(behavior_table_one_pair(weekly["WED"], ("SPY","QQQ"), day="WED"))

print("Table 4 Trading Time & Duration Statistics - WED | SPY–OIH pair")
display(table4)
```

In Table 7, I evaluate directional forecasting performance using a **sign success ratio**, following the definition in the original paper: the sample mean $S_{t+1}^{(m)}/n_1$ over the out-of-sample evaluation window. To quantify uncertainty under time dependence, I also compute a bootstrap-based standard error using the stationary bootstrap with a geometric block-length distribution.

```python
import numpy as np
import pandas as pd

def stationary_bootstrap_indices(n, mean_block_len=10, rng=None):
    if rng is None:
        rng = np.random.default_rng(123)
    p = 1.0 / float(mean_block_len)
    idx = np.empty(n, dtype=int)
    idx[0] = rng.integers(0, n)
    for t in range(1, n):
        if rng.random() < p:
            idx[t] = rng.integers(0, n)
        else:
            idx[t] = (idx[t-1] + 1) % n
    return idx

def stationary_bootstrap_se_of_mean(x, B=400, mean_block_len=10, seed=123):
    x = np.asarray(x, dtype=float)
    x = x[~np.isnan(x)]
    n = len(x)
    if n <= 2:
        return np.nan
    rng = np.random.default_rng(seed)
    means = np.empty(B, dtype=float)
    for b in range(B):
        idx = stationary_bootstrap_indices(n, mean_block_len=mean_block_len, rng=rng)
        means[b] = np.mean(x[idx])
    return float(np.std(means, ddof=1))

def sign_success_ratio_from_S(df_pair, yhat, n0, n1, realized_col="y"):
    ev = eval_index(df_pair, n0, n1)
    strat = strategy_returns_from_yhat(df_pair, yhat, realized_col=realized_col)
    s_vec = strat.loc[ev, "S_correct"].dropna().astype(float).values
    return float(np.mean(s_vec)) if len(s_vec) else np.nan, s_vec

def table7_for_pair_wed(weekly_wed, pair,mean_block_len=10, B=400, seed=123,
                        q_ma=1, c_q=0.5, pmax=20):
    i, j = pair
    df = build_pair_variables(weekly_wed, i, j)
    df = slice_pair_period(df, pair)
    n0 = 104
    n1 = N1_TABLE["WED"][pair]
    yhat_naive = forecast_naive_mean(df, n0=n0, n1=n1)
    yhat_ma    = forecast_arima_ma(df, q=q_ma, n0=n0, n1=n1)
    c = choose_c_fixed(df, train_end=n0, q=c_q)
    yhat_pw    = forecast_piecewise_linear(df, c=c, n0=n0, n1=n1)
    yhat_diff  = forecast_diff_logvol_ar(df, n0=n0, n1=n1, pmax=pmax)
    yhat_ratio = forecast_logvol_ratio_ar(df, n0=n0, n1=n1, pmax=pmax)

    models = {
        "Naïve":                 (yhat_naive, "y"),
        "Mov.Avg.":              (yhat_ma,    "y"),
        "Differ. of Volatility": (yhat_diff,  "v"),
        "Volatility Ratio":      (yhat_ratio, "v"),
        "Piecewise Linear":      (yhat_pw,    "y"),
    }
    ssr, se = {}, {}
    for name, (yh, target) in models.items():
        m, s_vec = sign_success_ratio_from_S(df, yh, n0=n0, n1=n1, realized_col=target)
        ssr[name] = m
        se[name]  = stationary_bootstrap_se_of_mean(
            s_vec, B=B, mean_block_len=mean_block_len, seed=seed
        )
    out = pd.DataFrame([ssr, se], index=["Sign success ratio", "Standard Error"])
    return out

def print_table7(mean_block_len=10, B=400, seed=123):
    print("Table 7 | Sign Success Ratio - Wednesday")
    for title, pair in [
        ("Panel A: SPY-OIH pair", ("SPY","OIH")),
        ("Panel B: SPY-XLF pair", ("SPY","XLF")),
        ("Panel C: SPY-QQQ pair", ("SPY","QQQ")),
    ]:
        tab = table7_for_pair_wed(
            weekly["WED"], pair,
            mean_block_len=mean_block_len, B=B, seed=seed
        )
        print(f"\n{title}")
        display(tab.round(3))

print_table7(mean_block_len=10, B=400, seed=123)
```

In Table 8, I present the results of **chi-square type tests** to examine the statistical dependence between trading signals and volatility regimes. This analysis directly tests Hypothesis 3, which posits that directional forecasting success is systematically related to volatility conditions.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def chi2_stat_2x2(a, b):
    a = np.asarray(a, dtype=int)
    b = np.asarray(b, dtype=int)
    tab = np.zeros((2, 2), dtype=int)
    for i in (0, 1):
        for j in (0, 1):
            tab[i, j] = int(np.sum((a == i) & (b == j)))
    if (tab.sum(axis=0) == 0).any() or (tab.sum(axis=1) == 0).any():
        return 0.0
    stat, _, _, _ = chi2_contingency(tab, correction=False)
    return float(stat)

def stationary_bootstrap_indices(n, mean_block_len=10, rng=None):
    if rng is None:
        rng = np.random.default_rng(123)
    p = 1.0 / float(mean_block_len)
    idx = np.empty(n, dtype=int)
    idx[0] = rng.integers(0, n)
    for t in range(1, n):
        if rng.random() < p:
            idx[t] = rng.integers(0, n)
        else:
            idx[t] = (idx[t - 1] + 1) % n
    return idx

def stationary_bootstrap_pvalue_chi2(a, b, B=400, mean_block_len=10, seed=123):
    df = pd.DataFrame({"a": a, "b": b}).dropna()
    a = df["a"].astype(int).values
    b = df["b"].astype(int).values
    n = len(a)
    if n <= 5:
        return np.nan
    obs = chi2_stat_2x2(a, b)
    rng = np.random.default_rng(seed)
    stats = np.empty(B, dtype=float)
    for k in range(B):
        ia = stationary_bootstrap_indices(n, mean_block_len=mean_block_len, rng=rng)
        ib = stationary_bootstrap_indices(n, mean_block_len=mean_block_len, rng=rng)
        stats[k] = chi2_stat_2x2(a[ia], b[ib])
    return float(np.mean(stats >= obs))

def build_YZX_on_eval(df_pair, yhat, ev_idx):
    d = df_pair.copy()
    yhat_al = yhat.reindex(d.index).astype(float)
    yhat_prev = yhat_al.loc[ev_idx]
    y_real    = d["y"].loc[ev_idx]
    x_relvol  = d["v"].loc[ev_idx]
    X = (x_relvol > 0).astype(int)
    Y = (yhat_prev > 0).astype(int)
    Z = ((yhat_prev > 0) & (y_real > 0)).astype(int)
    out = pd.DataFrame({"Y": Y, "Z": Z, "X": X}, index=ev_idx).dropna()
    return out

def table8_for_pair(weekly_by_day, pair, day, mean_block_len=10,
                    B=400, seed=123, n0=104, slice_func=slice_pair_period):
    i, j = pair
    df = build_pair_variables(weekly_by_day, i, j)
    df = slice_func(df, pair)
    n1 = N1_TABLE[day][pair]
    ev = eval_index(df, n0, n1)

    yhat_naive = forecast_naive_mean(df, n0=n0, n1=n1)
    yhat_ma    = forecast_arima_ma(df, q=1, n0=n0, n1=n1)
    c = choose_c_fixed(df, train_end=n0, q=0.3)
    yhat_pw    = forecast_piecewise_linear(df, c=c, n0=n0, n1=n1)
    yhats = {"Naïve": yhat_naive, "MA": yhat_ma, "Piecewise": yhat_pw}
    rows = {}
    for name, yhat in yhats.items():
        yzX = build_YZX_on_eval(df, yhat, ev_idx=ev)
        p_YX = stationary_bootstrap_pvalue_chi2(yzX["Y"], yzX["X"], B=B, mean_block_len=mean_block_len, seed=seed)
        p_ZX = stationary_bootstrap_pvalue_chi2(yzX["Z"], yzX["X"], B=B, mean_block_len=mean_block_len, seed=seed)
        rows[name] = {"Y,X": p_YX, "Z,X": p_ZX}
    return pd.DataFrame(rows).T

def table8_panel(pair, mean_block_len=10, B=400, seed=123, slice_func=slice_pair_period):
    out_cols = []
    out_data = []
    for day in ["MON", "WED", "FRI"]:
        t = table8_for_pair(
            weekly[day], pair, day, mean_block_len=mean_block_len,
            B=B, seed=seed, n0=104, slice_func=slice_func
        )
        out_cols += [(day, "Y,X"), (day, "Z,X")]
        out_data.append(t[["Y,X", "Z,X"]])
    wide = pd.concat(out_data, axis=1)
    wide.columns = pd.MultiIndex.from_tuples(out_cols)
    return wide

def print_table8(slice_func=slice_pair_period):
    print("Table 8 | Test for Independence")
    for title, pair in [
        ("Panel A: SPY-OIH pair", ("SPY", "OIH")),
        ("Panel B: SPY-XLF pair", ("SPY", "XLF")),
        ("Panel C: SPY-QQQ pair", ("SPY", "QQQ")),
    ]:
        wide = table8_panel(pair, mean_block_len=10, B=400, seed=123, slice_func=slice_func)
        wide = wide.reindex(columns=pd.MultiIndex.from_product([["MON","WED","FRI"], ["Y,X","Z,X"]]))
        print(f"\n{title}")
        display(wide.round(3))

print_table8()
```

To further assess robustness, the **extension analysis** re-tests all three hypotheses using data from 2020 to 2025. By applying the same empirical framework to a new time period, the extension provides an additional out-of-sample validation. The fact that rotation strategies **continue to outperform** benchmarks in most cases suggests that the original results are not purely driven by sample-specific effects.
<br><br>

Overall, the evidence indicates that while extreme overfitting is unlikely, moderate sample dependence remains. Predictability appears to be sector-specific and regime-dependent, implying that the economic significance of rotation strategies should be interpreted with caution when applied to new markets or periods.


# 10. Opportunities for Further Research

This study demonstrates that relative return and volatility predictability can be exploited through a binary asset rotation framework. However, several important research directions remain open.
<br><br>

First, future research may investigate whether nonlinear and machine learning models can improve forecasting performance. A natural hypothesis is that models such as random forests, neural networks, or regime-switching models can better capture structural changes and nonlinear dependencies in financial time series, leading to higher directional accuracy and improved portfolio performance. This hypothesis can be tested by replacing the linear forecasting models with machine learning algorithms and comparing out-of-sample returns, Sharpe ratios, and sign success ratios under the same rolling-window framework.
<br><br>

Second, the current strategy uses a fully-invested binary allocation rule. A potential hypothesis is that allowing partial allocation or risk-adjusted position sizing can improve risk-adjusted returns while reducing drawdowns. Future studies may implement volatility targeting, mean-variance optimization, or dynamic leverage constraints and evaluate their impact using portfolio simulation techniques.
<br><br>

Finally, the empirical analysis is limited to weekly ETF data from U.S. markets. A further hypothesis is that the rotation strategy exhibits different predictive power across asset classes, frequencies, and market environments. Future research may extend the dataset to include international equities, commodities, bonds, and higher-frequency data, and apply panel data methods to evaluate cross-market robustness.
<br><br>

Overall, these extensions would deepen the understanding of time-series predictability and enhance the practical relevance of asset rotation strategies.
